```python
national_team_players = fifa["nationality"].value_counts()
```

```python
national_team_players.head(10)
```

```python
plt.figure(figsize=(14,4))
plt.xlabel("Country Name")
plt.ylabel("No. Of Players Playing")
plt.title("Comparison of players plaing for their country")
plt.plot(national_team_players.head(10))
plt.show()
```

```python

keys = fifa["nationality"].value_counts()[0:10].keys()
value = fifa["nationality"].value_counts()[0:10]
plt.figure(figsize=(14,4))
plt.xlabel("Country Name")
plt.ylabel("No. Of Players Playing")
plt.title("Bar chart Representation of above graph")
plt.bar(list(fifa["nationality"].value_counts()[0:10].keys()), list(fifa["nationality"].value_counts()[0:10]))
plt.show()
```

```python
player_jerseys = fifa[["short_name", "team_jersey_number"]].head(10)
player_jerseys
```

```python
plt.figure(figsize= (17,5))
plt.title(" Players graph With highest Jersey number")
plt.xlabel("Player name")
plt.ylabel("Player Jersey Number")
player_jersey = player_jerseys.sort_values(by=["team_jersey_number"], ascending=False)
player_jersey
plt.bar(list(player_jersey["short_name"])[0:10], list(player_jersey["team_jersey_number"]), color ="b")
plt.show()
```
